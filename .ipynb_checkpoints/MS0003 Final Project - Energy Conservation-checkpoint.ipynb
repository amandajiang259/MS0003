```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


# Load the dataset (assuming 'owid-energy-data.csv' is in the same directory)
data=pd.read_csv('owid-energy-data.csv')

# --- Define the selected features ---
green_features = [
    'renewables_share_energy', 'renewables_electricity', 'solar_electricity',
    'wind_electricity', 'hydro_electricity', 'low_carbon_share_energy',
    'nuclear_electricity', 'renewables_energy_per_capita',
    'biofuel_electricity', 'other_renewable_electricity'
]

fossil_features = [
    'fossil_share_energy', 'fossil_electricity', 'coal_electricity',
    'oil_electricity', 'gas_electricity', 'coal_consumption',
    'oil_consumption', 'gas_consumption', 'fossil_fuel_consumption',
    'carbon_intensity_elec' # Higher carbon intensity means less green
]

all_selected_features = green_features + fossil_features

# --- Step 1: Filter data for the year 2020 ---
year_to_analyze = 2020
df_2020 = data[data['year'] == year_to_analyze].copy()

# Ensure only selected features and 'country' are kept for processing
df_2020_subset = df_2020[['country'] + all_selected_features]

# --- Step 2: Handle missing data by filling with the mean of each column ---
# Calculate means based on the entire dataset for robustness, then apply to 2020 subset
# Or, calculate means only from the 2020 subset if you prefer year-specific imputation
# For this draft, we'll use the mean from the 2020 subset for simplicity.
for col in all_selected_features:
    if col in df_2020_subset.columns:
        mean_value = df_2020_subset[col].mean()
        df_2020_subset[col].fillna(mean_value, inplace=True)
    else:
        print(f"Warning: Column '{col}' not found in 2020 subset. Skipping imputation.")

# --- Step 3 & 4: Min-Max Scaling and Inverting Fossil Features ---
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Create a DataFrame to store scaled values
df_scaled = pd.DataFrame(index=df_2020_subset.index)
df_scaled['country'] = df_2020_subset['country']

# Scale green features
for feature in green_features:
    if feature in df_2020_subset.columns:
        # Reshape for scaler (needs 2D array)
        scaled_values = scaler.fit_transform(df_2020_subset[[feature]])
        df_scaled[f'{feature}_scaled'] = scaled_values.flatten()
    else:
        print(f"Warning: Green feature '{feature}' not found. Skipping scaling.")

# Scale and invert fossil features
for feature in fossil_features:
    if feature in df_2020_subset.columns:
        # Reshape for scaler (needs 2D array)
        scaled_values = scaler.fit_transform(df_2020_subset[[feature]])
        # Invert the scale: 1 - scaled_value, so higher original value means lower 'green' score
        df_scaled[f'{feature}_scaled'] = (1 - scaled_values).flatten()
    else:
        print(f"Warning: Fossil feature '{feature}' not found. Skipping scaling and inversion.")

# --- Step 5: Combine into a single "Green Energy Index" metric ---
# Sum of scaled green features
df_scaled['green_score'] = df_scaled[[f'{f}_scaled' for f in green_features if f'{f}_scaled' in df_scaled.columns]].sum(axis=1)

# Sum of scaled (and inverted) fossil features
df_scaled['fossil_score'] = df_scaled[[f'{f}_scaled' for f in fossil_features if f'{f}_scaled' in df_scaled.columns]].sum(axis=1)

# Calculate the final Green Energy Index
# A higher index means greener.
df_scaled['Green_Energy_Index'] = df_scaled['green_score'] + df_scaled['fossil_score']

# Normalize the final index to be between 0 and 1 (optional, but good for interpretability)
# This step ensures the final index is also on a consistent scale, useful for comparison.
min_index = df_scaled['Green_Energy_Index'].min()
max_index = df_scaled['Green_Energy_Index'].max()
if (max_index - min_index) > 0:
    df_scaled['Green_Energy_Index_Normalized'] = (df_scaled['Green_Energy_Index'] - min_index) / (max_index - min_index)
else:
    df_scaled['Green_Energy_Index_Normalized'] = 0.5 # Handle case where all values are the same

# Display the top and bottom countries by Green Energy Index
result_df = df_scaled[['country', 'Green_Energy_Index_Normalized']].sort_values(
    by='Green_Energy_Index_Normalized', ascending=False
).reset_index(drop=True)

print(f"Green Energy Index for {year_to_analyze} (Normalized 0-1):")
print(result_df.head(20))
print("\n...")
print(result_df.tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# --- Define the selected features ---
green_features = [
    'renewables_share_energy', 'renewables_electricity', 'solar_electricity',
    'wind_electricity', 'hydro_electricity', 'low_carbon_share_energy',
    'nuclear_electricity', 'renewables_energy_per_capita',
    'biofuel_electricity', 'other_renewable_electricity'
]

fossil_features = [
    'fossil_share_energy', 'fossil_electricity', 'coal_electricity',
    'oil_electricity', 'gas_electricity', 'coal_consumption',
    'oil_consumption', 'gas_consumption', 'fossil_fuel_consumption',
    'carbon_intensity_elec'
]

all_selected_features = green_features + fossil_features

# --- Define years to process ---
start_year = 1990
end_year = 2022
years_to_analyze = range(start_year, end_year + 1)

# --- Initialize list to collect all yearly results ---
all_results = []

# --- Loop through each year ---
for year in years_to_analyze:
    df_year = data[data['year'] == year].copy()

    if df_year.empty:
        print(f"Skipping year {year}: no data.")
        continue

    # Select relevant columns
    df_year_subset = df_year[['country'] + all_selected_features].copy()

    # Fill missing values using mean of the year (fallback to global mean or 0)
    for col in all_selected_features:
        if col in df_year_subset.columns:
            mean_val = df_year_subset[col].mean(skipna=True)
            if pd.isna(mean_val):
                mean_val = data[col].mean(skipna=True)
            if pd.isna(mean_val):
                mean_val = 0
            df_year_subset[col] = df_year_subset[col].fillna(mean_val)

    # --- Scale and Invert ---
    scaler = MinMaxScaler()
    df_scaled = pd.DataFrame(index=df_year_subset.index)
    df_scaled['country'] = df_year_subset['country']
    df_scaled['year'] = year

    for feature in green_features:
        if feature in df_year_subset.columns:
            scaled = scaler.fit_transform(df_year_subset[[feature]])
            df_scaled[f'{feature}_scaled'] = scaled.flatten()

    for feature in fossil_features:
        if feature in df_year_subset.columns:
            scaled = scaler.fit_transform(df_year_subset[[feature]])
            df_scaled[f'{feature}_scaled'] = (1 - scaled).flatten()

    # --- Compute Index ---
    df_scaled['green_score'] = df_scaled[[f'{f}_scaled' for f in green_features if f'{f}_scaled' in df_scaled.columns]].sum(axis=1)
    df_scaled['fossil_score'] = df_scaled[[f'{f}_scaled' for f in fossil_features if f'{f}_scaled' in df_scaled.columns]].sum(axis=1)
    df_scaled['Green_Energy_Index'] = df_scaled['green_score'] + df_scaled['fossil_score']

    # Normalize the index per year
    min_val = df_scaled['Green_Energy_Index'].min()
    max_val = df_scaled['Green_Energy_Index'].max()
    if max_val > min_val:
        df_scaled['Green_Energy_Index_Normalized'] = (df_scaled['Green_Energy_Index'] - min_val) / (max_val - min_val)
    else:
        df_scaled['Green_Energy_Index_Normalized'] = 0.5

    # Collect results
    all_results.append(df_scaled[['country', 'year', 'Green_Energy_Index_Normalized']])

# --- Combine all years ---
result_df_all_years = pd.concat(all_results).reset_index(drop=True)

# --- Plotting ---
countries_to_plot = ['United States', 'China', 'Germany', 'India', 'Norway', 'Brazil', 'Canada', 'Australia', 'Japan', 'United Kingdom']
plot_df = result_df_all_years[result_df_all_years['country'].isin(countries_to_plot)]

plt.figure(figsize=(14, 8))
for country in countries_to_plot:
    country_data = plot_df[plot_df['country'] == country]
    plt.plot(country_data['year'], country_data['Green_Energy_Index_Normalized'], label=country, marker='o')

plt.title('Green Energy Index Over Time (Normalized 0-1)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Green Energy Index (Normalized)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(np.arange(start_year, end_year + 1, 2), rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Define the years for analysis ---
start_year = 2000
end_year = 2024
years_to_analyze = range(start_year, end_year + 1)

# List to store results for all years and countries
all_results = []

# --- Loop through each year to calculate the Green Energy Index ---
for year in years_to_analyze:
    print(f"Processing year: {year}")
    df_year = data[data['year'] == year].copy()

    # Ensure only selected features and 'country' are kept for processing
    # Also include 'renewables_share_energy' for direct plotting later
    df_year_subset = df_year[['country', 'renewables_share_energy'] + all_selected_features]

    # --- Handle missing data by filling with the mean of each column for the current year ---
    for col in all_selected_features:
        if col in df_year_subset.columns:
            # Calculate mean for the current year's data
            mean_val_current_year = df_year_subset[col].mean()
            
            # Determine the fill value
            # pd.isna() correctly handles scalar NaNs and Series that are effectively NaN (e.g., empty or all NaN)
            if pd.isna(mean_val_current_year): 
                # Fallback to the global mean from the entire dataset
                mean_val_global = data[col].mean()
                
                if pd.isna(mean_val_global): # If global mean is also NaN
                    fill_value = 0 # Final fallback
                else:
                    fill_value = mean_val_global
            else:
                fill_value = mean_val_current_year

            df_year_subset[col].fillna(fill_value, inplace=True)
        else:
            print(f"Warning: Column '{col}' not found in {year} subset. Skipping imputation.")

    # --- Min-Max Scaling and Inverting Fossil Features ---
    scaler = MinMaxScaler()

    df_scaled_year = pd.DataFrame(index=df_year_subset.index)
    df_scaled_year['country'] = df_year_subset['country']
    df_scaled_year['year'] = year
    df_scaled_year['renewables_share_energy'] = df_year_subset['renewables_share_energy'] # Keep original for plotting

    # Scale green features
    for feature in green_features:
        if feature in df_year_subset.columns and not df_year_subset[feature].isnull().all():
            scaled_values = scaler.fit_transform(df_year_subset[[feature]])
            df_scaled_year[f'{feature}_scaled'] = scaled_values.flatten()
        else:
            # If column is all NaN or missing, fill scaled value with 0.5 (neutral)
            df_scaled_year[f'{feature}_scaled'] = 0.5
            print(f"Warning: Green feature '{feature}' in {year} is all NaN or missing. Filling scaled value with 0.5.")

    # Scale and invert fossil features
    for feature in fossil_features:
        if feature in df_year_subset.columns and not df_year_subset[feature].isnull().all():
            scaled_values = scaler.fit_transform(df_year_subset[[feature]])
            # Invert the scale: 1 - scaled_value, so higher original value means lower 'green' score
            df_scaled_year[f'{feature}_scaled'] = (1 - scaled_values).flatten()
        else:
            # If column is all NaN or missing, fill scaled value with 0.5 (neutral)
            df_scaled_year[f'{feature}_scaled'] = 0.5
            print(f"Warning: Fossil feature '{feature}' in {year} is all NaN or missing. Filling scaled value with 0.5.")

    # --- Combine into a single "Green Energy Index" metric ---
    # Sum of scaled green features
    df_scaled_year['green_score'] = df_scaled_year[[f'{f}_scaled' for f in green_features if f'{f}_scaled' in df_scaled_year.columns]].sum(axis=1)

    # Sum of scaled (and inverted) fossil features
    df_scaled_year['fossil_score'] = df_scaled_year[[f'{f}_scaled' for f in fossil_features if f'{f}_scaled' in df_scaled_year.columns]].sum(axis=1)

    # Calculate the final Green Energy Index
    df_scaled_year['Green_Energy_Index'] = df_scaled_year['green_score'] + df_scaled_year['fossil_score']

    # Normalize the final index to be between 0 and 1 for the current year's data
    min_index = df_scaled_year['Green_Energy_Index'].min()
    max_index = df_scaled_year['Green_Energy_Index'].max()
    if (max_index - min_index) > 0:
        df_scaled_year['Green_Energy_Index_Normalized'] = (df_scaled_year['Green_Energy_Index'] - min_index) / (max_index - min_index)
    else:
        df_scaled_year['Green_Energy_Index_Normalized'] = 0.5 # Handle case where all values are the same

    all_results.append(df_scaled_year[['country', 'year', 'Green_Energy_Index_Normalized', 'renewables_share_energy']])

# Concatenate all yearly results into a single DataFrame
final_results_df = pd.concat(all_results).reset_index(drop=True)

print("\n--- Sample of Calculated Green Energy Index over Years ---")
print(final_results_df.head())
print("\n--- Plotting Green Energy Index and Renewables Share Energy ---")

# --- Plotting ---
# Select a few countries for plotting to avoid clutter.
# You can change this list to plot different countries.
countries_to_plot = ['United States', 'China', 'Germany', 'India', 'Norway', 'Brazil', 'Canada', 'Australia', 'Japan', 'United Kingdom']

# Filter data for selected countries
plot_df = final_results_df[final_results_df['country'].isin(countries_to_plot)]

# Create the plots
fig, axes = plt.subplots(2, 1, figsize=(14, 12), sharex=True)
fig.suptitle('Green Energy Index and Renewables Share Energy (2000-2024)', fontsize=16)

# Plot Green Energy Index
for country in countries_to_plot:
    country_data = plot_df[plot_df['country'] == country]
    axes[0].plot(country_data['year'], country_data['Green_Energy_Index_Normalized'], label=country, marker='o', markersize=4, linewidth=1.5)

axes[0].set_ylabel('Green Energy Index (Normalized)', fontsize=12)
axes[0].set_title('Green Energy Index Over Time', fontsize=14)
axes[0].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[0].set_ylim(0, 1) # Ensure y-axis is between 0 and 1

# Plot Renewables Share Energy
for country in countries_to_plot:
    country_data = plot_df[plot_df['country'] == country]
    axes[1].plot(country_data['year'], country_data['renewables_share_energy'], label=country, marker='o', markersize=4, linewidth=1.5)

axes[1].set_xlabel('Year', fontsize=12)
axes[1].set_ylabel('Renewables Share Energy (%)', fontsize=12)
axes[1].set_title('Renewables Share Energy Over Time', fontsize=14)
axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
axes[1].grid(True, linestyle='--', alpha=0.7)

# Set x-axis ticks to show all years clearly
axes[0].set_xticks(np.arange(start_year, end_year + 1, 2)) # Show every other year for readability
axes[0].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 0.85, 0.96]) # Adjust layout to make space for suptitle and legend
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset (assuming 'owid-energy-data.csv' is in the same directory)
data = pd.read_csv('owid-energy-data.csv')

# --- Define the selected features ---
green_features = [
    'renewables_share_energy', 'renewables_electricity', 'solar_electricity',
    'wind_electricity', 'hydro_electricity', 'low_carbon_share_energy',
    'nuclear_electricity', 'renewables_energy_per_capita',
    'biofuel_electricity', 'other_renewable_electricity'
]

fossil_features = [
    'fossil_share_energy', 'fossil_electricity', 'coal_electricity',
    'oil_electricity', 'gas_electricity', 'coal_consumption',
    'oil_consumption', 'gas_consumption', 'fossil_fuel_consumption',
    'carbon_intensity_elec' # Higher carbon intensity means less green
]

all_selected_features = green_features + fossil_features

# --- Define the years for analysis ---
start_year = 2000
end_year = 2024
years_to_analyze = range(start_year, end_year + 1)

# List to store results for all years and countries
all_results = []

# --- Loop through each year to calculate the Green Energy Index ---
for year in years_to_analyze:
    print(f"Processing year: {year}")
    df_year = data[data['year'] == year].copy()

    # Ensure only selected features and 'country' are kept for processing
    # Also include 'renewables_share_energy' for direct plotting later
    df_year_subset = df_year[['country', 'renewables_share_energy'] + all_selected_features]

    # --- Handle missing data by filling with the mean of each column for the current year ---
    for col in all_selected_features:
        if col in df_year_subset.columns:
            # Calculate mean for the current year's data and safely get the scalar value
            mean_val_current_year = df_year_subset[col].mean()
            if isinstance(mean_val_current_year, pd.Series):
                mean_val_current_year = mean_val_current_year.item() # Extract scalar from Series of length 1

            # Determine the fill value
            if pd.isna(mean_val_current_year): # This should now be a scalar check
                # Fallback to the global mean from the entire dataset
                mean_val_global = data[col].mean()
                if isinstance(mean_val_global, pd.Series):
                    mean_val_global = mean_val_global.item()

                if pd.isna(mean_val_global): # If global mean is also NaN
                    fill_value = 0 # Final fallback
                else:
                    fill_value = mean_val_global
            else:
                fill_value = mean_val_current_year

            df_year_subset[col].fillna(fill_value, inplace=True)
        else:
            print(f"Warning: Column '{col}' not found in {year} subset. Skipping imputation.")

    # --- Min-Max Scaling and Inverting Fossil Features ---
    scaler = MinMaxScaler()

    df_scaled_year = pd.DataFrame(index=df_year_subset.index)
    df_scaled_year['country'] = df_year_subset['country']
    df_scaled_year['year'] = year
    df_scaled_year['renewables_share_energy'] = df_year_subset['renewables_share_energy'] # Keep original for plotting

    # Scale green features
    for feature in green_features:
        if feature in df_year_subset.columns and not df_year_subset[feature].isnull().all():
            scaled_values = scaler.fit_transform(df_year_subset[[feature]])
            df_scaled_year[f'{feature}_scaled'] = scaled_values.flatten()
        else:
            # If column is all NaN or missing, fill scaled value with 0.5 (neutral)
            df_scaled_year[f'{feature}_scaled'] = 0.5
            print(f"Warning: Green feature '{feature}' in {year} is all NaN or missing. Filling scaled value with 0.5.")

    # Scale and invert fossil features
    for feature in fossil_features:
        if feature in df_year_subset.columns and not df_year_subset[feature].isnull().all():
            scaled_values = scaler.fit_transform(df_year_subset[[feature]])
            # Invert the scale: 1 - scaled_value, so higher original value means lower 'green' score
            df_scaled_year[f'{feature}_scaled'] = (1 - scaled_values).flatten()
        else:
            # If column is all NaN or missing, fill scaled value with 0.5 (neutral)
            df_scaled_year[f'{feature}_scaled'] = 0.5
            print(f"Warning: Fossil feature '{feature}' in {year} is all NaN or missing. Filling scaled value with 0.5.")

    # --- Combine into a single "Green Energy Index" metric ---
    # Sum of scaled green features
    df_scaled_year['green_score'] = df_scaled_year[[f'{f}_scaled' for f in green_features if f'{f}_scaled' in df_scaled_year.columns]].sum(axis=1)

    # Sum of scaled (and inverted) fossil features
    df_scaled_year['fossil_score'] = df_scaled_year[[f'{f}_scaled' for f in fossil_features if f'{f}_scaled' in df_scaled_year.columns]].sum(axis=1)

    # Calculate the final Green Energy Index
    df_scaled_year['Green_Energy_Index'] = df_scaled_year['green_score'] + df_scaled_year['fossil_score']

    # Normalize the final index to be between 0 and 1 for the current year's data
    min_index = df_scaled_year['Green_Energy_Index'].min()
    max_index = df_scaled_year['Green_Energy_Index'].max()
    if (max_index - min_index) > 0:
        df_scaled_year['Green_Energy_Index_Normalized'] = (df_scaled_year['Green_Energy_Index'] - min_index) / (max_index - min_index)
    else:
        df_scaled_year['Green_Energy_Index_Normalized'] = 0.5 # Handle case where all values are the same

    all_results.append(df_scaled_year[['country', 'year', 'Green_Energy_Index_Normalized', 'renewables_share_energy']])

# Concatenate all yearly results into a single DataFrame
final_results_df = pd.concat(all_results).reset_index(drop=True)

print("\n--- Sample of Calculated Green Energy Index over Years ---")
print(final_results_df.head())
print("\n--- Plotting Green Energy Index and Renewables Share Energy ---")

# --- Plotting ---
# Select a few countries for plotting to avoid clutter.
# You can change this list to plot different countries.
countries_to_plot = ['United States', 'China', 'Germany', 'India', 'Norway', 'Brazil', 'Canada', 'Australia', 'Japan', 'United Kingdom']

# Filter data for selected countries
plot_df = final_results_df[final_results_df['country'].isin(countries_to_plot)]

# Create the plots
fig, axes = plt.subplots(2, 1, figsize=(14, 12), sharex=True)
fig.suptitle('Green Energy Index and Renewables Share Energy (2000-2024)', fontsize=16)

# Plot Green Energy Index
for country in countries_to_plot:
    country_data = plot_df[plot_df['country'] == country]
    axes[0].plot(country_data['year'], country_data['Green_Energy_Index_Normalized'], label=country, marker='o', markersize=4, linewidth=1.5)

axes[0].set_ylabel('Green Energy Index (Normalized)', fontsize=12)
axes[0].set_title('Green Energy Index Over Time', fontsize=14)
axes[0].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[0].set_ylim(0, 1) # Ensure y-axis is between 0 and 1

# Plot Renewables Share Energy
for country in countries_to_plot:
    country_data = plot_df[plot_df['country'] == country]
    axes[1].plot(country_data['year'], country_data['renewables_share_energy'], label=country, marker='o', markersize=4, linewidth=1.5)

axes[1].set_xlabel('Year', fontsize=12)
axes[1].set_ylabel('Renewables Share Energy (%)', fontsize=12)
axes[1].set_title('Renewables Share Energy Over Time', fontsize=14)
axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
axes[1].grid(True, linestyle='--', alpha=0.7)

# Set x-axis ticks to show all years clearly
axes[0].set_xticks(np.arange(start_year, end_year + 1, 2)) # Show every other year for readability
axes[0].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 0.85, 0.96]) # Adjust layout to make space for suptitle and legend
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# --- Step 1: Prepare data ---
feature_2020 = data[data['year'] == 2020][[
    'country',
    'gdp',
    'population',
    'energy_per_capita',
    'electricity_demand',
    'electricity_demand_per_capita',
    'per_capita_electricity',
    'primary_energy_consumption',
    'fossil_share_energy',
    'renewables_share_energy',
    'carbon_intensity_elec',
    'oil_share_energy',
    'gas_share_energy',
    'coal_share_energy',
    'low_carbon_share_energy',
    'greenhouse_gas_emissions',
    'net_elec_imports_share_demand',
    'electricity_share_energy'  # target
]].dropna()

# Define features and target
X = feature_2020.drop(columns=['country', 'electricity_share_energy'])
y = feature_2020['electricity_share_energy']

# --- Step 2: Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# --- Step 3: Scale features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Step 4: Train model ---
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# --- Step 5: Evaluate ---
y_pred = model.predict(X_test_scaled)
print("Model performance on test set:")
print(f"R² score: {r2_score(y_test, y_pred):.3f}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred):.3f}")

# --- Step 6: Predict for a new country ---
# Example: Suppose you have a dictionary with new country feature values
new_country_features = {
    'gdp': 50000,
    'population': 10000000,
    'energy_per_capita': 150,
    'electricity_demand': 120000000,
    'electricity_demand_per_capita': 12000,
    'per_capita_electricity': 4500,
    'primary_energy_consumption': 800000000,
    'fossil_share_energy': 0.6,
    'renewables_share_energy': 0.3,
    'carbon_intensity_elec': 0.4,
    'oil_share_energy': 0.2,
    'gas_share_energy': 0.25,
    'coal_share_energy': 0.15,
    'low_carbon_share_energy': 0.25,
    'greenhouse_gas_emissions': 400,
    'net_elec_imports_share_demand': 0.05
}

# Convert to DataFrame and scale
import numpy as np
new_country_df = pd.DataFrame([new_country_features])
new_country_scaled = scaler.transform(new_country_df)

# Predict electricity_share_energy
predicted_share = model.predict(new_country_scaled)
print(f"Predicted electricity share of primary energy: {predicted_share[0]:.3f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# --- 1. Load dataset ---
data = pd.read_csv('owid-energy-data.csv')
data = data[(data['iso_code'].notna()) & (data['iso_code'].str.len() == 3)]


```

```python
# --- 2. Create Clusters Based on Energy Mix Trends (2000-2020) ---
data_cluster = data[data['year'].isin([2000, 2020])].copy()

def get_feature_change(df, feature):
    pivot = df.pivot(index='country', columns='year', values=feature)
    pivot = pivot.dropna(subset=[2000, 2020])
    change = (pivot[2020] - pivot[2000]) / pivot[2000].replace(0, np.nan)
    return change

features_for_clustering = [
    'fossil_share_energy',
    'renewables_share_energy',
    'greenhouse_gas_emissions',
    'carbon_intensity_elec'
]

feature_changes = pd.DataFrame()
for f in features_for_clustering:
    feature_changes[f + '_change'] = get_feature_change(data_cluster, f)

feature_changes = feature_changes.dropna()

scaler = StandardScaler()
X_cluster = scaler.fit_transform(feature_changes)

kmeans = KMeans(n_clusters=4, random_state=42)
feature_changes['cluster'] = kmeans.fit_predict(X_cluster)
feature_changes = feature_changes.reset_index()  # country becomes a column
```

```python
# --- 3. Prepare Supervised Learning Dataset (2020 features → predict 2025) ---
df_past = data[data['year'] == 2020]
df_future = data[data['year'] == 2025][['country', 'renewables_share_energy']].dropna()

# Merge 2020 with 2025 target
df = pd.merge(df_past, df_future, on='country', suffixes=('_2020', '_2025'))

# Merge cluster labels
df = pd.merge(df, feature_changes[['country', 'cluster']], on='country')
```

```python


# --- 4. Select Features and Target ---
features = [
    'fossil_share_energy_2020',
    'renewables_share_energy_2020',
    'carbon_intensity_elec_2020',
    'greenhouse_gas_emissions_2020',
    'gdp_2020',
    'population_2020',
    'electricity_demand_2020',
    'cluster'
]
target = 'renewables_share_energy_2025'

# Clean data
df_model = df[features + [target]].dropna()
X = df_model[features]
y = df_model[target]

```

```python
# --- 5. Train-Test Split + Scaling ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42
)

# --- 6. Train Model ---
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- 7. Evaluate ---
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n📊 Model Performance:")
print(f"R² Score: {r2:.3f}")
print(f"MSE: {mse:.3f}")

# --- 8. Plot Results ---
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.8)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Renewable Share (2025)')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted Renewable Share (2025)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
